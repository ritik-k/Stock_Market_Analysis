```python
stock = {}  # creating empty dict to store dataframes
stock_list = ['BKNG','EXPE','ALGT','LYV','SIX','FUN','DIS','PENN','DAL','UAL','AAL','LUV','ALK','AC','CCL','RCL','NCLH','LVS','MGM','WYNN','CZR','ERI','MAR','HLT','H','CHH','WH','PK','MTN','VAC']
# ^ creating list of all stock tickers
```

```python
# writing a function that will make a separate dataframe for each stock.
for df in stock_list:
    symbol = df                      
    data_source='yahoo'
    start_date = '2020-01-01'
    end_date = '2020-04-30'
    stock[df] = data.DataReader(symbol, data_source, start_date, end_date)  # storing the resulting df in dict
    stock[df] = stock[df][['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] # repositioning columns for convenience
```

```python
def all_adj_c(x):
    adj_c_19 = x.loc['2020-02-19']['Adj Close']
#     return adj_c_19
    adj_c_24 = x.loc['2020-03-24']['Adj Close']
#     return adj_c_24
    adj_c_p = ((adj_c_19 - adj_c_24)/adj_c_19)*100
    return adj_c_19, adj_c_24, adj_c_p
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=stock['BKNG'].index, y=stock['BKNG']['Adj Close'],
                    mode='lines+markers',
                    name='Booking Holdings'))
fig.add_trace(go.Scatter(x=stock['EXPE'].index, y=stock['EXPE']['Adj Close'],
                    mode='lines+markers',
                    name='Expedia Group'))
fig.add_trace(go.Scatter(x=stock['ALGT'].index, y=stock['ALGT']['Adj Close'],
                    mode='lines+markers', name='Allegiant Travel'))

fig.update_layout(shapes=[
    dict(
      type= 'line',
      yref= 'paper', y0= 0, y1= 1,
      xref= 'x', x0= '2020-02-19', x1= '2020-02-19'
    ),
])

fig.show()
```

```python
def plot_f(s1,n1,s2,n2,s3,n3):
    import plotly.graph_objects as go
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=s1.index, y=s1['Adj Close'],mode='lines+markers',name=n1))
    fig.add_trace(go.Scatter(x=s2.index, y=s2['Adj Close'],mode='lines+markers',name=n2))
    fig.add_trace(go.Scatter(x=s3.index, y=s3['Adj Close'],mode='lines+markers',name=n3))
    fig.update_layout(shapes=[
    dict(
      type= 'line',
      yref= 'paper', y0= 0, y1= 1,
      xref= 'x', x0= '2020-02-19', x1= '2020-02-19'),])
    fig.show()
    
```

```python
plot_f(stock['BKNG'],'Booking Holdings',stock['EXPE'],'Expedia Group',stock['ALGT'],'Allegiant Travel')
```
